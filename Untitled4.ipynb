```python
import pandas as pd 
import os
```

```python
train_raw = pd.read_csv("rice_challenge/Train.csv")
test_raw = pd.read_csv("rice_challenge/Test.csv")

image_folder = 'rice_challenge/images/'
image_path_col = "filepath"

train_raw.insert(loc=1, column=image_path_col, value= image_folder+train_raw["Image_id"])
test_raw.insert(loc=1, column=image_path_col, value= image_folder+test_raw["Image_id"])

train = train_raw[train_raw.apply(lambda row: os.path.exists(row[image_path_col]), axis=1)]
test = test_raw[test_raw.apply(lambda row: os.path.exists(row[image_path_col]), axis=1)]

print(f"[INFO] Missing images in dataset for Train : { train_raw.shape[0]-train.shape[0]} " )
print(f"[INFO] Missing images in dataset for Test : { test_raw.shape[0]-test.shape[0]} " )
```

```python
from fastai.vision.all import *
```

```python
out = Image.new('RGB', (600, 300))
out.paste(Image.open('rice_challenge/images/id_00vl5wvxq3.jpg').resize((300, 300)), (0, 0))
out.paste(Image.open('rice_challenge/images/id_00vl5wvxq3_rgn.jpg').resize((300, 300)), (300, 0))
out
```

```python
train_rgb = train[~train.Image_id.str.contains('_rgn')] # Just the RGB images
dls = ImageDataLoaders.from_df(train_rgb, fn_col='Image_id', label_col='Label', 
                               folder='rice_challenge/images', item_tfms=Resize(400),
                               batch_tfms = aug_transforms(size=224))
dls.show_batch() # Note the augmentation applied to the images!
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_top_losses(9, figsize=(15,10))
```

```python
interp.plot_confusion_matrix()
```

```python
test_rgb= test[~test.Image_id.str.contains('_rgn')] # Just the RGB images
preds, _ = learn.get_preds(dl=dls.test_dl(test_rgb)) # Getting the predicted probabilities
preds.shape
```
